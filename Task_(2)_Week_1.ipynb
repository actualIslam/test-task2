# Tasks

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Task (2) Dataset.csv')
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
sort_by_city = df['City'].value_counts()
sort_by_city
```

```python
grouped = df.groupby('City')
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TrafficIndexLive', bins=20, kde=True)
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df['Datetime'])
plt.title('Distribution of Datetime')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='JamsCount', y='JamsLength')
plt.title('Relationship between JamsCount and JamsLength')
plt.xlabel('JamsCount')
plt.ylabel('JamsLength')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(10, 6))
plt.pie(sort_by_city.values, labels=sort_by_city.index, autopct='%1.1f%%', radius=1.5)
plt.title('Distribution of City')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='TrafficIndexLive', data=df)
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
```
